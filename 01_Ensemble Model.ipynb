```python
import pandas as pd
import numpy as np
import random
from random import randint
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split, ShuffleSplit
from sklearn.datasets import make_regression
from autogluon.tabular import TabularDataset, TabularPredictor
from datetime import datetime
```

```python
#materialData = pd.read_excel('Results_Impute/materialData_ii_imputed.xlsx')
materialData = pd.read_excel('Results_Impute/materialData_ii_imputed_text_numeric.xlsx')
materialData
```

```python
rs = ShuffleSplit(n_splits=1, test_size=0.25, random_state=0)  # n_splits=1 means we're doing one split

for train_index, test_index in rs.split(materialData):
    train_data = materialData.iloc[train_index]  
    test_data =materialData.iloc[test_index]
    
train_data.to_csv('trainESM.csv', index=False) 
test_data.to_csv('testESM.csv', index=False)   
```

```python
train_data.head()
```

```python
label = 'materialDensity'
train_data[label].describe()
```

```python
predictor = TabularPredictor(label=label, problem_type='regression').fit(train_data, 
                                                                         presets='best_quality', 
                                                                         hyperparameters='multimodal',#train more models
                                                                         time_limit=3600)
```

```python
results = predictor.fit_summary()
```

```python
test_data = TabularDataset('testESM.csv')
```

```python
Y_test = test_data['materialDensity']
```

```python
test_data.head()
```

```python
y_pred = predictor.predict(test_data.drop(columns=[label]))
y_pred.head()
print(y_pred)
```

```python
combineY = pd.DataFrame ({
    'Y_test' : Y_test,
    'y_pred' : y_pred})

combineY.to_excel("Result_Enemble Model.xlsx")
```

```python
predictor.evaluate(test_data, silent=True)
```

```python
leaderboard = predictor.leaderboard(test_data, silent=True)

for index, row in leaderboard.iterrows():
    print(f"Model {index+1}:")
    print(row.to_dict())  # This prints the row details for each model
```
