```python
import pandas as pd
import numpy as np
from io import StringIO
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
materialData = pd.read_excel('Results_Impute/materialData_ii_imputed.xlsx')
materialData
```

## K fold Split

```python
#Define dependent and independent variables
##reshape(-1,1) reshapes the column to 1D, 2D array
Y = np.array(materialData.materialDensity).reshape(-1,1)
X = np.array(materialData[['materialConductivity','materialHeatCapacity']])
```

```python
from sklearn.model_selection import KFold
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

#Print the ShuffleSplit object:
kf.get_n_splits(X)
kf.get_n_splits(Y)

#Generate and print train/test indices for each split:
#multiple feature matrices (e.g., X1,X2,X3), combine to a single feature matrix before splitting the data into training and testing sets. 
for i, (train_index, test_index) in enumerate(kf.split(X, Y)):
    print(f"Fold {k}:")
    print(f"  Train: index={train_index}")
    print(f"  Test:  index={test_index}")
    
    Y_train, Y_test = Y[train_index], Y[test_index]
    X_train, X_test = X[train_index], X[test_index]
    
    print(f"  X_train: \n{X_train}")
    print(f"  X_test: \n{X_test}")
    print(f"  Y_train: {Y_train}")
    print(f"  Y_test: {Y_test}")
```

```python
from sklearn.linear_model import LinearRegression 
model = LinearRegression()
```

```python
from sklearn.model_selection import cross_val_score 
scores = cross_val_score(model, X, Y, cv=kf, scoring='r2')
```

```python
import numpy as np 

average_r2 = np.mean(scores) 

print(f"R² Score for each fold: {[round(score, 4) for score in scores]}")
print(f"Average R² across {k} folds: {average_r2:.2f}")
```

## Extra Tree Regressor

max_depth=2 Meaning: Limits the number of splits each tree can make to 3 layers. Importance: Prevents overfitting by stopping trees from becoming too detailed, ensuring generalissaion to new data.

n_estimators=3: Meaning: Specifies the number of trees (detectives) in the ensemble. Importance: More trees reduce individual biases, resulting in a more robust average decision and less likelihood of overfitting.

max_features=3: Limits: the number of features each tree can consider at each split to a maximum of 3. Importance, introduce randomness, ensuring diverse trees and reducing overfitting by precenting any single feature from domincating decisions.

min_sample_split=2: Meanig: Sets the minimum number of samples required in a group to allow a split. Importance: ensure that splits are based on sufficient data, precenting the model from making unrealibale decision with small groups.

random_state=0

random forset: bootstraps the dataset for each tree

extra tree utilizes the entire dataset furint trainign

```python
from sklearn.ensemble import ExtraTreesRegressor
reg = ExtraTreesRegressor(n_estimators=5, max_depth=6, random_state=0, min_samples_split=2).fit(
   X_train, Y_train.ravel()) #ravel() convert a multi-dimensional array into a one-dimensional array
```

```python
reg.score(X_test, Y_test)
```

```python
predicted_Y_for_X_test = reg.predict(X_test)
print(predicted_Y_for_X_test)
```

```python
print('Actual values:', Y_test)
```

```python
y_predictions_flat = predicted_Y_for_X_test.flatten()
y_test_flat = Y_test.flatten() #flatten() creates a new copy of the data in a fresh one-dimensional array

material_conductivity = X_test[:, 0]  
material_heat_capacity = X_test[:, 1] 

output_df = pd.DataFrame({
    'Material Conductivity': material_conductivity,
    'Material Heat Capacity': material_heat_capacity,
    'Actual': y_test_flat,
    'Predicted': predictions_flat
})

# Export to CSV with proper precision
output_df.to_excel('Prediction_ExtraTreeRegressor.xlsx', float_format='%.5f', index=False)
```

```python
## some
```

p-value: The test rejects the hypothesis of normality when the p-value is less than or equal to 0.05. Failing the normality test allows you to state with 95% confidence the data does not fit the normal distribution. Passing the normality test only allows you to state no significant departure from normality was found. The observed effect (or feature importance) is highly statistically significant, with an extremely low probability that the result is due to random chance.

```python
# Error Analysis: Compute residuals
residuals = y_test_flat - y_predictions_flat

# Plot residuals vs. predicted values
plt.scatter(y_predictions_flat, residuals)
plt.title('Residuals vs. Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()

# Plot residuals distribution
plt.hist(residuals, bins=7)
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Perform a normality test on residuals
statistic, p_value = stats.normaltest(residuals)
print(f'Normality Test p-value: {p_value}')

#Q-Q Plot
stats.probplot(residuals, dist="norm", plot=plt)
plt.title('Q-Q Plot of Residuals')
plt.show()
```

## MAE
