# Data ingestion and preperation

We first feed in the data ("materialData_ii_imputed_text_numeric.xlsx")
Then there are two steps:

a) encoding and splitting the data for ML pipeline: Regression (ridge) and Extra Tree Regressor

b) not encode, splitting the data for ensemble model.

#### Material Data imputed

```python
#materialData = pd.read_excel('Results_Impute/materialData_ii_imputed.xlsx')
materialData = pd.read_excel('Results_Impute/materialData_ii_imputed_text_numeric_2.xlsx')
materialData
```

#### Encode ASHRAE for Lin Reg and Extra Tree

```python
encoder = OneHotEncoder(sparse_output=False)

#encode materialCategory
encode_materialData_materialCategory = pd.DataFrame(
    encoder.fit_transform(np.array(materialData.materialCategory).reshape(-1,1)),
    index = materialData.index)

# #encode materialDescription
# encode_materialData_materialDescription = pd.DataFrame(
#     encoder.fit_transform(np.array(materialData.materialDescription).reshape(-1,1)),
#     index = materialData.index)

#encode materialType
encode_materialData_materialType = pd.DataFrame(
    encoder.fit_transform(np.array(materialData.materialType).reshape(-1,1)),
    index = materialData.index)
```

```python
encode_materialData = pd.concat([
    encode_materialData_materialCategory,
    encode_materialData_materialType,
    materialData.materialConductivity,
    materialData.materialHeatCapacity,
    materialData.materialDensity
], axis=1)

print(len(encode_materialData))
```

#### ShuffelSplit (with encoding)

```python
#Previous splitting
rs = ShuffleSplit(n_splits=1, test_size=0.25, random_state=0)  # n_splits=1 means we're doing one split

for encode_train_index, encode_test_index in rs.split(encode_materialData):
    encode_train_data = encode_materialData.iloc[encode_train_index]
    encode_test_data = encode_materialData.iloc[encode_test_index]

#encode_train_data.to_csv('encodeTrain.csv', index=False) 
#encode_test_data.to_csv('encodeTest.csv', index=False)   
```

```python
from sklearn.model_selection import train_test_split

# First, split into training (70%) and temp (30%)
encode_train_data, temp_data = train_test_split(encode_materialData, test_size=0.30, random_state=0)

# Now, split the temp_data into validation (15%) and test (15%)
encode_vald_data, encode_test_data = train_test_split(temp_data, test_size=0.50, random_state=0)  # 50% of 30% is 15%

# Save the datasets to CSV files
#encode_train_data.to_csv('encodeTrain.csv', index=False)
#encode_vald_data.to_csv('encodeVald.csv', index=False)
#encode_test_data.to_csv('encodeTest.csv', index=False)

print("Data split completed: 70% training, 15% validation, 15% test.")
```

### Correlation Matrix 

There is moderate negative linear relationship between the two variables.
As material conductivity increases, material heat capacity tends to decrease, and vice versa.
The relationship is not perfect (which would be -1), but it's strong enough to suggest a meaningful inverse trend.

```python
correlation_matrix = encode_test_data[['materialConductivity', 'materialHeatCapacity', 'materialDensity']].corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

#### ShuffelSplit (no encoding)

```python
rs = ShuffleSplit(n_splits=1, test_size=0.25, random_state=0)  # n_splits=1 means we're doing one split

for train_index, test_index in rs.split(materialData):
    train_data = materialData.iloc[train_index]  
    test_data =materialData.iloc[test_index]
    
train_data.to_csv('train.csv', index=False) 
test_data.to_csv('test.csv', index=False)   
```

### Ensemble Model Predict individual material density

```python
def predict_material_ensemble():
    print("\nEnter material properties for Ensemble Prediction:")
    
    # Collect inputs
    category = input("Material Category: ")
    type = input("Material Type: ")
    conductivity = float(input("Thermal Conductivity (W/m·K): "))
    heat_capacity = float(input("Specific Heat Capacity (J/kg·K): "))
    material_description = input("describe briefly the material")

    # 1. Create a dictionary with names matching your original CSV/DataFrame columns
    user_data = {
        'materialCategory': [category],
        'materialType': [type],
        'materialConductivity': [conductivity], # Use the exact column names from your train_data
        'materialHeatCapacity': [heat_capacity],
        'materialDescription' : [material_description]
    }

    # 2. Convert to DataFrame (AutoGluon requires a DF or Series)
    input_df = pd.DataFrame(user_data)

    # 3. Predict using the TabularPredictor (no manual encoding needed!)
    prediction = predictor.predict(input_df)

    print(f"\nEnsemble Predicted Density: {prediction.iloc[0]:.2f} kg/m³")

# Run the function
predict_material_ensemble()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import mean_squared_error

# List of different tree sizes to track performance
n_estimators_range = [5, 10, 20, 50, 100]
train_errors = []
val_errors = []

for n in n_estimators_range:
    # Train model with different tree sizes
    model = ExtraTreesRegressor(n_estimators=n, max_depth=6, random_state=0, min_samples_split=10)
    model.fit(x_train_encode, y_train_encode.values.ravel())

    # Predict on training & validation sets
    y_train_pred = model.predict(x_train_encode)
    y_val_pred = model.predict(x_val_encode)

    # Compute Mean Squared Error (MSE)
    train_mse = mean_squared_error(y_train_encode, y_train_pred)
    val_mse = mean_squared_error(y_val_encode, y_val_pred)

    train_errors.append(train_mse)
    val_errors.append(val_mse)

# --- Visualization ---
plt.plot(n_estimators_range, train_errors, label="Train MSE", marker='o')
plt.plot(n_estimators_range, val_errors, label="Validation MSE", marker='o')
plt.xlabel("Number of Trees (n_estimators)")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("Train vs Validation Loss")
plt.legend()
plt.show()
```
