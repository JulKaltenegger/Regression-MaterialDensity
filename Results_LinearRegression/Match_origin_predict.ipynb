```python
MaterialData = pd.read_excel('materialDataToBeMatched.xlsx')
MaterialData
```

```python
# Initialize an empty list to store matched results
matched_results = []

# Loop through each row in the predictions DataFrame
for _, pred_row in MaterialData.iterrows():
    # Adjust `Y_test` to match the `Y-test` column in the loaded data
    y_test_value = pred_row['Y-test']
    
    # Find matching rows where `materialDensity` matches `Y_test`
    matching_rows = MaterialData[MaterialData['materialDensity'] == y_test_value]
    
    # Print the Y_test value and the corresponding materialDensity for comparison
    print(f"Y_test: {y_test_value}, matching materialDensity values:")
    
    if not matching_rows.empty:
        for _, orig_row in matching_rows.iterrows():
            matched_results.append({
                'materialCategory': orig_row['materialCategory'],
                'materialDescription': orig_row['materialDescription'],
                'materialDensity': orig_row['materialDensity'],
                'Y_test': pred_row['Y-test'],
                'y_linpred': pred_row['y_linpred']
            })
            # Print the matching materialDensity and materialCategory
            print(f" - Material Density: {orig_row['materialDensity']} (Category: {orig_row['materialCategory']}, Description: {orig_row['materialDescription']})")
    else:
        print(" - No matching materialDensity found.")

# Convert matched_results to a DataFrame for further analysis
matched_df = pd.DataFrame(matched_results)

# Remove duplicates from matched_df
matched_df = matched_df.drop_duplicates()

# Optionally, specify a subset of columns to check for duplicates if needed
# matched_df = matched_df.drop_duplicates(subset=['materialCategory', 'materialDescription', 'Y_test', 'Y_predict'])

# Display the resulting DataFrame after removing duplicates
print(matched_df)

# Save the matched DataFrame to an Excel file
matched_df.to_excel('match.xlsx', index=False)
```
